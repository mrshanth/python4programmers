## Functions in Python

```python
from __future__ import print_function
from IPython.display import Image

def do_nothing():  # no params
    pass

def do_nothing2():
    return

def do_nothing3():
    return None

do_nothing() == do_nothing2() == do_nothing3() == None
```

```python
def fib(n):  # one (positional) param
   if n == 1:
      return 1
   elif n == 0:   
      return 0            
   else:                      
      return fib(n-1) + fib(n-2)
```

```python
for i in range(1,10):
    print(i, fib(i))
```

```python
fib(-2)  # maximum recursion depth exceeded
```

```python
# multiple positional parameter
def add(x, y):
    return x + y

a = add(5, 6)
b = add(x=5, y=6)
c = add(y=6, x=5)
d = add(5, y=6)
a == b == c
```

```python
# optional (keyword) argument

def increment(x, step=1):
    return x + step

a = increment(10)
b = increment(5, 2)
c = increment(10, step=10)

all([a == 11, b == 7, c == 20]) # functional style 'and'
```

```python
# All parameters (arguments) in the Python language are passed by reference.

mylist = [1,2,3]
def addfour(anylist):
    anylist.append(4)

print(4 in mylist)
addfour(mylist)
print(4 in mylist)
```

```python
# never do this!!!!

def create_dict_with_kv(key, value, defaults={}):
    """ Returns a new dict with preset key and value """ # this is a docstring, aka. help
    defaults[key] = value
    return defaults

mydict = create_dict_with_kv(u'alma', u'körte')
myotherdict = create_dict_with_kv(u'banán', u'dinnye')
    
```

```python
print(mydict)
```

```python
print(myotherdict)
```

```python
mydict is myotherdict
```

```python
def fixed_create_dict_with_kv(key, value, defaults=None):
    d = {}
    if defaults is not None:
        d.update(defaults)
    d[key] = value
    return d

mydict = fixed_create_dict_with_kv(u'alma', u'körte')
myotherdict = fixed_create_dict_with_kv(u'banán', u'dinnye')
mythirddict = fixed_create_dict_with_kv(1,2, {'c': 4, 'd': 5})
```

```python
print(mydict, myotherdict, mythirddict)
```

```python
# magic arguments

def accept_any_arg(*args, **kwargs):
    print('args:', args, 'kwargs:', kwargs)
```

```python
accept_any_arg(1,2,3)
accept_any_arg(a=10)
accept_any_arg()
accept_any_arg('kutya', {}, y=0)
```

```python
def one_arg(x, *a, **kw):
    print('x:', x, 'args:', a, 'kwargs:', kw )

one_arg(5)
one_arg('d')
one_arg(5, 10, y=3)
one_arg(x=5, y=3)
```

### Beágyazás

```python
# nested functions

global_variable = True

def outer_function(x, y):
    def inner_function(z):
        global_variable = False
        print("inner function: called with {0}".format(z))
        print("inner function: i see x is {0}, y is {1}".format(x, y))
        print("inner_function: global_variable is {0}".format(global_variable))
    print("outer_function: global_variable is {0}".format(global_variable))
    return inner_function(x + y)
```

```python
outer_function('vaká', 'ció')
print('global_variable is', global_variable)
```

### Még érdekesebb dolgokat is lehet csinálni...

```python
def incrementator(x):
    def _increment(y):
        return x + y
    return _increment  # hoppa, hoppa, mi tortenik itt?

increment_with_six = incrementator(6)
increment_with_six(4) == 10
```

```python
increment_with_minusone = incrementator(-1)
increment_with_minusone(4)    
```

### Dekorátorok

```python
# 'keep everything' decorator
def i_am_a_decorator(func):
    return func

```

```python
supernewfunction = i_am_a_decorator(increment_with_minusone)
```

```python
supernewfunction(4)
```

```python
@i_am_a_decorator
def meaningoflife(x):
    return x - 42
```

```python
meaningoflife(42)
```

```python
def print_params(func):
    def wrapper(*args, **kwargs):
        print('function {0} called with {1}, {2}.format'.format(func, args, kwargs))
        return func(*args, **kwargs)
    return wrapper

@print_params
def mysum(_list):
    if len(_list) == 1:
        return _list[0]
    else:
        return _list[0] + mysum(_list[1:])
    
mysum([1,2,3])
```

### Egy összetett példa...

```python
@print_params
def myreduce(_list, _func):
    if len(_list) == 1:
        return _list[0]
    else:
        return _func(_list[0], myreduce(_list[1:], _func))

myreduce([1,2,3,4], lambda x, y : x + y)
```

```python
myreduce([1,2,3,4], lambda x, y : x * y)
```
